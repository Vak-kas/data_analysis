1907년 ~ 2018년 동안 서울의 기온데이터를 히스토그램으로 표현한 결과

```python
import csv
import matplotlib.pyplot as plt;
file_path = (os.path.join(colab_path, "seoul.csv"))
with open (file_path, encoding = "cp949") as f:
    data = csv.reader(f);
    next(data);
    result = [];


    for row in data:
        if row[-1] !="":
            result.append(float(row[-1]));

plt.hist(result, bins= 100, color = "r");
plt.show();
```

8월 데이터만 뽑아서 히스토그램으로 나타낸 결과

```python
import csv
import matplotlib.pyplot as plt;
file_path = (os.path.join(colab_path, "seoul.csv"))
with open (file_path, encoding = "cp949") as f:

    data = csv.reader(f);
    (next(data))
    aug = [];

    for row in data:
        if row[-1] !="" and row[0].split("-")[1] == "08":
            aug.append(float(row[-1]))

plt.hist(aug, bins = 100, color="r");
plt.show();
```

1월과 8월의 데이터를 히스토그램으로 시각화 하기

```python
import csv
import matplotlib.pyplot as plt;
file_path = (os.path.join(colab_path, "seoul.csv"))
with open (file_path, encoding = "cp949") as f:

    data = csv.reader(f);
    (next(data))
    aug = [];
    jan = []

    for row in data:
        if row[-1] !="":
            if row[0].split("-")[1] == "08":
                aug.append(float(row[-1]))
            if row[0].split("-")[1] == "01":
                jan.append(float(row[-1]));


plt.hist(aug, bins = 100, color="r", label = "Aug");
plt.hist(jan, bins = 100, color = "b", label = "Jan");
plt.legend();
plt.show();
```
