#데이터 시각화 하기

```python
import csv
import matplotlib.pyplot as plt;

file_path = (os.path.join(colab_path, "seoul.csv"))
with open(file_path, encoding = "cp949") as f:
    data = csv.reader(f);
    next(data);

    result = [];


    for row in data:
        if row[-1] !="":
            result.append(float(row[-1]));


plt.plot(result, "r"); #약 4만개의 데이터를 기온에 따라 그래프를 "r"색, 즉 빨간색으로 찍음
# plt.figure(figsize = (10, 2)); #가로로 10인치, 세로로 2인치로 설정
plt.show();
```

# 날짜 데이터 추출하기


1년 중 여름의 정점인 8월의 최고 기온 데이터만 추출해서 그래프로 그려보기(색깔은 핫핑크색)


```python
import csv
import matplotlib.pyplot as plt;

file_path = (os.path.join(colab_path, "seoul.csv"))
with open(file_path, encoding = "cp949") as f:
    data = csv.reader(f);
    next(data);

    result = [];


    for row in data:
        if row[-1] !="" and row[0].split("-")[1] == "08":
            result.append(float(row[-1]));


plt.plot(result, "hotpink"); #약 4만개의 데이터를 기온에 따라 그래프를 "r"색, 즉 빨간색으로 찍음
# plt.figure(figsize = (10, 2)); #가로로 10인치, 세로로 2인치로 설정
plt.show();
```

매년 내 생일 3월 7일의 최고 기온 데이터를 추출하여 그래프로 나타내기

```python
import csv
import matplotlib.pyplot as plt;

file_path = (os.path.join(colab_path, "seoul.csv"))
with open(file_path, encoding = "cp949") as f:
    data = csv.reader(f);
    next(data);

    result = [];


    for row in data:
        if row[-1] !="" and row[0].split("-")[1] == "03" and row[0].split("-")[2] == "07":
            result.append(float(row[-1]));
plt.plot(result, "hotpink"); #약 4만개의 데이터를 기온에 따라 그래프를 "r"색, 즉 빨간색으로 찍음
# plt.figure(figsize = (10, 2)); #가로로 10인치, 세로로 2인치로 설정
plt.show();
```

1983년 이후 데이터만 추출해서 그래프로 그려보기

```python
import csv
import matplotlib.pyplot as plt;

file_path = (os.path.join(colab_path, "seoul.csv"))
with open(file_path, encoding = "cp949") as f:
    data = csv.reader(f);
    next(data);

    high = [];
    low = [];


    for row in data:
        if row[-1] !="" and row[-2] !="" and row[0].split("-")[1] == "03" and row[0].split("-")[2] == "07":
            if int(row[0].split("-")[0][1:]) >= 1983:
                high.append(float(row[-1]));
                low.append(float(row[-2]));





plt.plot(high, "hotpink"); #high 리스트에 저장된 값을 hotpink색으로 그리기
plt.plot(low ,"skyblue"); #low 리스트에 저장된 값을 skyblue 색으로 그리기
# plt.figure(figsize = (10, 2)); #가로로 10인치, 세로로 2인치로 설정
plt.show(); #그래프 나타내기
```

```python
import csv
import matplotlib.pyplot as plt;
# !apt-get -qq -y install fonts-nanum*


file_path = (os.path.join(colab_path, "seoul.csv"))
with open(file_path, encoding = "cp949") as f:
    data = csv.reader(f);
    next(data);

    high = [];
    low = [];


    for row in data:
        if row[-1] !="" and row[-2] !="" and row[0].split("-")[1] == "03" and row[0].split("-")[2] == "07":
            if int(row[0].split("-")[0][1:]) >= 1983:
                high.append(float(row[-1]));
                low.append(float(row[-2]));

plt.rc('font', family='DejaVu Sans')
plt.rcParams["axes.unicode_minus"] = False

# 그래프 크기 조정
plt.figure(figsize=(10, 5))

plt.plot(high, "hotpink", label = "high"); #high 리스트에 저장된 값을 hotpink색으로 그리기
plt.plot(low ,"skyblue", label = "low"); #low 리스트에 저장된 값을 skyblue 색으로 그리기
# plt.figure(figsize = (10, 2)); #가로로 10인치, 세로로 2인치로 설정
plt.title("High and low temperature graph for my birthday March 7th")
plt.legend();
plt.show(); #그래프 나타내기
```
