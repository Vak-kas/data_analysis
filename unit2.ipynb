# mathplotlib 라이브러리 사용하기


1. import matplotlib.pyplot as plt 로 함수불러오기
2. plt.plot([x축 데이터], [y축 데이터])
3. plt.show() 로 그래프 보여주기

```
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40])
plt.show();
```

```
import matplotlib.pyplot as plt
plt.plot([1, 2 ,3, 4], [12, 43, 25, 15])
plt.show

```

그래프에 옵션 추가하기

1. plt.title("제목") : 제목 넣기

```
import matplotlib.pyplot as plt
plt.title("plotting");
plt.plot([10, 20, 30, 40]);
plt.show();
```

그래프 옵션 추가하기
2. 그래프에 범례 넣기
plt.lenged(loc = 위치)


2    9     1

6    10      5

3    8     4

```
import matplotlib.pyplot as plt
plt.title("legend");
plt.plot([10, 20, 30, 40], label="asc");
plt.plot([40, 30, 20, 10], label="desc");
plt.legend();
plt.show();
```

그래프 옵션 추가하기
3. 그래프 색상 바꾸기
plot(color = "색상")
r, g, b, k(검정), y(노랑)....등등 있음

```
import matplotlib.pyplot as plt;
plt.title("color");
#그래프 그리기
plt.plot([10, 20, 30, 40], color = "skyblue", label="skyblue");
plt.plot([40, 30, 20, 10], color="pink", label="pink");
plt.legend();
plt.show();
```

그래프 옵션 추가하기
4. 그래프 선 모양 바꾸기
plot(linestyle="--") 또는 plot(ls=":")


```
import matplotlib.pyplot as plt
plt.title("linestyle");
#빨간색 dash그래프
plt.plot([10, 20, 30, 40], color = "red", ls= "--", label="dashed");
#초록색 dotted 그래프
plt.plot([40, 30, 20, 10], color="g", ls=":", label="dotted");

plt.legend();
plt.show();
```

그래프 옵션 추가하기
5. 마커 모양 바꾸기
선 형태가 아닌, 점 형태의 그래프를 그릴 수 있음.
r. = 점, g^ = 삼각형 마커



```
import matplotlib.pyplot as plt;
plt.title("marker");
plt.plot([10, 20, 30, 40], "r.", label="circle");
plt.plot([40, 30, 20, 10], "g^", label="triangle up");
plt.legend();
plt.show();
```
